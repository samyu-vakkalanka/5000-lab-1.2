# Baking Desserts
## Skill Level: Difficult
### Macarons
Certain desserts are known for requiring more skills. Macarons are one of these recipes. There are a lot of components that impact how these desserts turn out and making the perfect macaron takes skill, practice, and patience. The crusts need to be crispy, with soft cookies and an even coloring. 

### Swirl Cakes
These cakes are also known as a difficult dessert to make. It requires baking a cake to the perfect thickness, letting it cool to the right temperature before frosting it, and rolling the cake carefully without breaking it. Again, with practice and patience, this beautiful dessert is possible.

```python
import matplotlib.pyplot as plt
x = list(range(10))
y = [2^a for a in x]
plt.plot(x,y)
plt.xlabel('Hours to bake')
plt.ylabel('Difficulty of dessert')
plt.title('Change in time required to bake dessert based on difficulty level')
```
